## Data PreProcessing

### Handling Missing Values

```python
# Let's the plot the distribution of this feature and based on that decide how to impute the missing values.

sns.histplot(df['Cholesterol'], kde=True)
plt.axvline(x=df['Cholesterol'].mean(), linestyle='-', color='red', label='mean')
plt.axvline(x=df['Cholesterol'].median(), linestyle='--', color='red', label='median')
plt.legend()
plt.show()

print('-'*72)
print(f'Skewness : {df['Cholesterol'].skew()}')
print(f'Kurtosis : {df['Cholesterol'].kurtosis()}')
```

The 'Cholesterol' column is not following a normal distribution so imputing it with mean might not be the best idea.
But yet the standard deviation technique can be used to impute the values with a range of high probability.<br>
The following things can be tried :<br>
1. Impute with median
2. Impute with mean
3. Impute with a range of values with high probability (ex : [mean-std, mean+std])

```python
# Number of missing values in testing and training.
n_missing_train = train['Cholesterol'].isnull().sum()
n_missing_test = test['Cholesterol'].isnull().sum()

# Mean, Median and std of cholesterol in training set.
mean = train['Cholesterol'].mean()
median = train['Cholesterol'].median()
std = train['Cholesterol'].std()

# Range of high probability values for imputation.
high_prob_values_train = np.random.randint(low=mean-std, high=mean+std, size=n_missing_train)
high_prob_values_test = np.random.randint(low=mean-std, high=mean+std, size=n_missing_test)
```

```python
train['Cholesterol'] = train['Cholesterol'].replace({np.nan:mean})
test['Cholesterol'] = test['Cholesterol'].replace({np.nan:mean})
```

```python
# There's only one value of RestingBP missing in train set.
train['RestingBP'] = train['RestingBP'].replace({np.nan:train['RestingBP'].mean()})
```
